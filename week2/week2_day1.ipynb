### 2-5. 데이터 필터링
1. 단일조건으로 데이터 선택
2. 여러 조건을 사용한 데이터 선택
3. 특정 값을 포함하는 데이터 선택
4. 특정 조건을 만족하는 행의 인덱스만 선택

```python
data = {
    '학생' : ['철수', '짱구', '훈이', '유리', '수지', '흰둥이', '짱아', '맹구', '미리', '성아'],
    '국어' : [88, 92, 76, 81, 95, 67, 80, 91, 85, 73],
    '영어' : [78, 85, 92, 88, 76, 84, 90, 81, 89, 80],
    '수학' : [90, 78, 82, 94, 87, 79, 83, 88, 76, 92],
    '사회' : [85, 92, 88, 77, 91, 82, 90, 79, 84, 88],
    '과학' : [91, 79, 88, 85, 82, 90, 87, 94, 83, 86]
}
df = pd.DataFrame(data)
df
```

```python
df[df['국어'] >= 80]
```

```python
# 국어 점수가 80점 이상이고, 사회 점수가 85점인 데이터
# and 조건 '&'
df[(df['국어'] >= 80) & (df['사회'] == 85)]
```

```python
# 영어 점수가 85점 이상이거나, 수학 점수가 90점 이상인 데이터
# or 조건 '|'
df[(df['영어'] >= 85) | (df['수학'] >= 90)]
```

```python
# 과학 점수가 91점이거나 86점인 데이터
df[df['과학'].isin([91, 86])]
```

```python
# 영어 점수가 90점 이상인 데이터의 인덱스만 알고 싶다
df[df['영어'] >= 90].index
```

### 3-1. 결측치 처리
- fillna : 결측치 채우기

```python
data_with_nan = {
    '고객ID': [101, 102, 103, 104, 105],
    '이름': ['수연' ,'서윤', '소리', '시은', None],
    '나이': [27, None, 24, 27, 26],
    '구매금액': [25000, 30000, None, 50000, 100000],
    '구매횟수': [1, 2, 1, 4, 5]
}

df_with_nan = pd.DataFrame(data_with_nan)
df_with_nan
```

```python
df_with_nan.isnull()
```

```python
# 각 컬럼 별로 결측치가 몇 개 있는지?
df_with_nan.isnull().sum()
```

```python
# 총 결측치가 몇 개 있는지?
df_with_nan.isnull().sum().sum()
```

#### 결측치 처리 방법
1. 결측치가 포함된 행 제거
2. 결측치가 포함된 열 제거
3. 결측치를 특정 값으로 대체 (ex. 0)
4. 결측치를 평균값으로 대체

```python
# 결측치가 포함된 행 제거
# inplace=True 혹은 다른 이름의 변수로 저장해줘야 삭제 상태가 저장이 된다
df_drop_row = df_with_nan.dropna()
df_drop_row
```

```python
# 결측치가 포함된 열 제거
# axis=0은 행, axis=1은 열
df_drop_column = df_with_nan.dropna(axis=1)
df_drop_column
```

```python
# 결측치를 채우자
# float으로 변환된다
df_fill_zero = df_with_nan.fillna(0)
df_fill_zero
```

```python
# 결측치를 평균 값으로
df_with_nan['나이'].fillna(df_with_nan['나이'].mean(), inplace=True)
df_with_nan
```

```python
df_with_nan['구매금액'].fillna(df_with_nan['구매금액'].mean(), inplace=True)
df_with_nan
```

```python
# 문자열 결측치를 채우는 방법? 알수없음, 미상 등으로 할 수 있다
```

3. 데이터 타입 변환

시간 열의 문자열 값(예: '118분')을 정수형으로 변환해 영화 시간(분)을 나타내도록 변경하세요.

```python
movie_fill['시간'] = movie_fill['시간'].str.replace('분', '').astype('int')
movie_fill
```

```python
new_time = []
for i in movie_fill['시간']:
    i = i.replace('분', '')
    new_time.append(int(i))

movie_fill['시간'] = new_time
movie_fill
```

6. 컬럼명 수정 및 정렬

장르 열을 ‘영화장르’로, 시간 열을 ‘상영시간’으로 수정하세요.  
평점을 기준으로 내림차순 정렬하고, 상위 10개의 영화를 출력하세요.

```python
movie_rename = movie_fill.rename(columns={'장르': '영화장르', '시간': '상영시간'})
movie_rename
```

```python
movie_sort = movie_rename.sort_values(by='평점', ascending=False)
movie_sort = movie_sort.reset_index(drop=True)
movie_sort.head(10)
```

7. 중복값 제거

영화 제목에 중복된 값이 있는지 확인하고, 중복된 영화가 있다면 제거하세요.

```python
movie_sort.duplicated(subset=['제목'])
```

```python
movie_dropd = movie_sort.drop_duplicates(subset=['제목'])
movie_dropd
```

8. 그룹별 통계

장르별로 평점의 평균을 계산하고, 각 장르별 평균 평점을 출력하세요.  
groupby를 사용하여 특정 열에 대한 통계 값을 구하세요.

```python
movie_dropd.groupby('영화장르')['평점'].mean()
```

```python
movie_dropd.groupby('국가')['상영시간'].mean()
```

9. 데이터 병합과 결합

새로운 영화 데이터를 추가하여 concat()을 사용해 기존 데이터프레임에 병합하세요.  
merge()를 활용해 각 영화에 감독 이름과 같은 새로운 정보를 추가해 보세요 (필요시 가상 데이터를 생성해 시도해 볼 수 있습니다).

```python
new_data = pd.DataFrame({
    '제목': ['영화A', '영화B'],
    '영화장르': ['액션', '드라마'],
    '국가': ['대한민국', '미국'],
    '상영시간': [110, 95],
    '평점': [8.5, 7.9]
})
```

```python
movie_con = pd.concat([movie_dropd, new_data], ignore_index=True)
movie_con
```

```python
director = {
    '제목': ['정국 : 아이 엠 스틸', '임영웅 | 아임 히어로 더 스타', '브레드이발소 : 빵스타의 탄생', '룩백', '비긴 어게인', '소년시절의 너', '에이리언 : 로물루스', '사랑의 하츄핑'],
    '영화감독': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
}

director_df = pd.DataFrame(director)
director_df
```

```python
movie_merge = pd.merge(movie_con, director_df, on='제목')
movie_merge
```
